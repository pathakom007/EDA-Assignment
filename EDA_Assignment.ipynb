2.  Explain overfitting and underfitting in ML. How can you detect and prevent
them?


1. Overfitting in Machine Learning:
   - Overfitting happens when a model learns noise, random fluctuations, and very specific patterns in the training data → it performs very well on training data but poorly on unseen (test) data.

   How to Detect Overfitting:
     - Training accuracy >> Test accuracy
     - Cross-validation shows instability (high variance across folds)

   How to Prevent Overfitting:
     - Regularization
        - L1 (Lasso) → feature selection (sparse models)
        - L2 (Ridge) → reduces weights, stabilizes model
        - Elastic Net → combination of L1 + L2
     - Reduce Model Complexity
        - Limit tree depth (Decision Tree)
        - Reduce polynomial degree (Polynomial Regression)
     - Cross-Validation
        - k-fold CV helps tune hyperparameters & avoid overfitting

2. Underfitting in Machine Learning:
   - Underfitting happens when a model is too simple to learn the relationships in data → it performs poorly on both training and test data.

   How to Detect Underfitting:
     - Training and test errors are both high
     - Both learning curves level off at high error

     How to Prevent Underfitting:
       - Increase model complexity
           - Use deeper trees
           - Higher polynomial degree
           - More layers/neurons in DL

       - Train longer (for neural networks)
       - Feature engineering
           - Add new features
           - Extract more meaningful features

    Bias–Variance Tradeoff:
       - This is fundamental to understanding overfitting and underfitting.

       Bias -
          - Error due to overly simplistic assumptions

       Variance  -
          - Error due to too much sensitivity to training data

    Cross-Validation (CV):
       - A technique to evaluate model performance reliably.

       Why Use CV:
          - Detect overfitting
          - Tune hyperparameters

       How CV Helps:
          -  If performance varies widely across folds → high variance (overfitting).
          - If performance is consistently bad → high bias (underfitting).

    Regularization Techniques:

          1. L1 Regularization (Lasso)

             - Adds |weights| penalty

             - Pushes weights to zero → feature selection

             - Useful when many useless features exist

         2. L2 Regularization (Ridge)

             - Adds weights² penalty

             - Prevents weights from growing large

             - Stabilizes model → reduces variance

         3. Elastic Net

             - Combines L1 + L2

             - Good for highly correlated features

5.  Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.

  - Feature scaling is a critical preprocessing step in ML because many algorithms perform poorly when input features have very different scales.

     - Why Is Feature Scaling Important in Machine Learning?:

       - Many ML algorithms are distance-based or gradient-based.
        If features have very different ranges, the model becomes biased toward features with larger numeric values.

  1. Min–Max Scaling (Normalization):
  
     - Formula - X(scaled) = X - X(min)/ X(max) - X(min)

     Output Range:

     - Typically [0, 1]
     - Sometimes [-1, 1] variant is used

     When to use:

     - When distribution is not Gaussian
     - When you want bounded values
     - Best for neural networks, deep learning, KNN, SVM

   2. Standardization (Z-score Scaling):

     - Formula - X(scaled) = X - mean/ standard deviation   

     Output Range:

      - Not bounded
      - Mean = 0
      - Standard deviation = 1

      When to use:

      - When data has outliers
      - When algorithm assumes normal distribution
      - Works best for linear regression, logistic regression, SVM, K-Means, PCA

      Impact on Distance-Based Algorithms:

6. Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?

1. Lable Encoding:
  
  - Label Encoding converts categories into integer labels.
    
     When to Use:
       - Use when categories have an ORDINAL relationship (meaning order matters).

      Examples of ordinal categories:
       - Education level → {Primary < High School < Graduate < Masters < PhD}
       - Size → {Small < Medium < Large}
       - Customer rating → {Bad < Average < Good < Excellent}

     Why:
       - Label Encoding preserves order (0 < 1 < 2).

       Do NOT use Label Encoding for nominal variables:
          - Red = 0
          - Blue = 1
          - Green = 2
       - Green(2) > Blue(1) > Red(0) → wrong & misleading.

       Algorithms where Label Encoding is harmful:
          - Linear Regression
          - Logistic Regression
          - SVM
          - KNN

2. One-Hot Encoding:

  - Creates a binary column for each category.

     When to Use:
       - Use for nominal (non-ordered) categories.

      Examples of nominal categories:

        - Gender → {Male, Female, Other}
        - City → {Delhi, Mumbai, Kolkata}
        - Color → {Red, Blue, Green}
        - Product Type → {Phone, Laptop, Tablet}

      Why:
        - No false ordering
        - Treats each category independently

      Algorithms where One-Hot Encoding is ideal:
         - Linear Regression, Logistic Regression
         - KNN (distance-based)
         - SVM
         - Neural Networks

         Ordinal vs. Nominal → The Key Decision

Which categories tend to have higher vs lower ratings — and why:

  From prior research and broad observations of app stores:

   - Apps in utility/productivity / essential-service categories tend to have higher average ratings. Because they’re functional: if they work well (good performance, stability, usefulness), users tend to reward them.

   - Niche or specialized categories (e.g. health, finance, certain enterprise tools) may also see high ratings — but only if they satisfy core needs well.

   - On the other hand, entertainment-heavy categories (e.g. some games, social, lifestyle) can show more variance in ratings. Some apps may delight users → high rating; others may disappoint → low rating. Hence average may not always be very high.

   - Categories prone to flooding with many low-quality or spammy apps (often games, utilities with ad-ware) may have lower average ratings.

   - Also, categories with many free apps, or apps with many “casual” users, might have low commitment — leading to more negative reviews (bugs, ads, performance issues), dragging down rating.

```python
import pandas as pd

# Load the dataset directly from your GitHub repository
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/master/googleplaystore.csv"
df = pd.read_csv(url)

# Clean dataset
df = df[['Category', 'Rating']].dropna()          # remove missing ratings/categories
df = df[df['Rating'] <= 5]                        # remove invalid rating values
df = df[df['Rating'] >= 1]                        # remove invalid low values

# Compute average rating per category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# Print results
print("=== Average Rating per Category ===")
print(category_ratings)

print("\n=== Highest Rated Categories (Top 5) ===")
print(category_ratings.head(5))

print("\n=== Lowest Rated Categories (Bottom 5) ===")
print(category_ratings.tail(5))
```

   3. When 10–5 days are left → noticeable price spike.
      - Business travelers book late → airlines charge more.

   4. Last 3–1 days → VERY HIGH exponential surge (2x to 5x).
      - Airlines maximize revenue from last-minute fliers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load dataset from the GitHub repo (use raw link)
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"
df = pd.read_csv(url)

print("First few rows:")
print(df.head())
print("\nColumns:")
print(df.columns.tolist())
print("\nShape:", df.shape)
print("\nInfo:")
print(df.info())

# 2. Quick summary of price
print("\nPrice statistics:")
print(df['price'].describe())

# 3. If the dataset has 'Days_Left' column --> use it; else compute if you have journey/search date
# Corrected: Changed 'Days_Left' to 'days_left' to match the actual column name
if 'days_left' in df.columns:
    df_days = df[['days_left','price']].dropna()
else:
    # This part will now only be reached if 'days_left' (lowercase) is genuinely missing
    raise ValueError("Dataset has no 'days_left' column — need column specifying how many days before departure booking was made.")

# Remove invalid or negative days (if any)
df_days = df_days[df_days['days_left'] >= 0]

# 4. Part (a): Price vs Days Left
avg_price_by_day = df_days.groupby('days_left')['price'].mean().reset_index().sort_values('days_left')
print("\nAverage price by days_left (first 10 rows):")
print(avg_price_by_day.head(10))

plt.figure(figsize=(10,5))
sns.lineplot(data=avg_price_by_day, x='days_left', y='price', marker='o')
plt.title("Average Flight Price vs Days Left Until Departure")
plt.xlabel("Days Left until Travel")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()

# 5. Window-based summary (choose some windows)
def window_price(df, low, high):
    mask = (df['days_left'] >= low) & (df['days_left'] <= high)
    return df.loc[mask, 'price'].mean()

windows = [(30,60), (20,30), (10,20), (5,10), (0,5)]
result = []
for low, high in windows:
    result.append({'window': f"{low}-{high}", 'avg_price': window_price(df_days, low, high)})
print("\nAverage price by booking window:")
print(pd.DataFrame(result))

# 6. Part (b): Compare airlines on a route (if route info exists)
if {'airline', 'source', 'destination'}.issubset(df.columns):
    route = ('Delhi', 'Mumbai')
    route_df = df[(df['source']==route[0]) & (df['destination']==route[1])]
    print(f"\nNumber of flights for route {route[0]} → {route[1]}:", route_df.shape[0])
    if not route_df.empty:
        stats = route_df.groupby('airline')['price'].agg(['mean','median','count']).sort_values('mean')
        print("\nAirline price stats on this route:")
        print(stats)
        plt.figure(figsize=(8,4))
        sns.barplot(data=stats.reset_index(), x='airline', y='mean')
        plt.xticks(rotation=45, ha='right')
        plt.title(f"Avg Price by Airline: {route[0]} → {route[1]}")
        plt.ylabel("Average Price")
        plt.tight_layout()
        plt.show()
    else:
        print("No flights found for this route in data.")
else:
    print("\nRoute or airline columns not present — cannot compare airlines for route.")
```
